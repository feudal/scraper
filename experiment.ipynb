```python
import requests
from bs4 import BeautifulSoup

r=requests.get("https://librarius.md/ro/books?sortBy=denumirea_asc&perPage=100")
#https://librarius.md/ro/books/page/2?sortBy=denumirea_asc&perPage=100
c=r.content

base_url=[]
begin_url="https://librarius.md/ro/books/page/"
end_url="?sortBy=denumirea_asc&perPage=100"
for page in range(2,443):
    base_url.append(begin_url+str(page)+end_url)
    
for url in range(2,443):
    r=requests.get(base_url[url])
    c.append(r)

#---------------------------------------------------------------------------------

soup=BeautifulSoup(c,"html.parser")

#print(soup.prettify())

#len(all)

all=soup.find_all("div",{"class":"book-card-content"})

all[0].find("a",{"class":"book-card-title-link"}).text
```

```python
for item in all:
    print(item.find("a",{"class":"book-card-title-link"}).text)
    print(item.find("div",{"class":"book-card-price-current"}).text.strip())
    try:
        print(item.find("span",{"class":"book-card-author-link"}).text.strip())
    except:
        print(None)
    print(" ")
```

```python
l=[]
for item in all:
    d={}
    d["Title"]=(item.find("a",{"class":"book-card-title-link"}).text)
    d["Price"]=(item.find("div",{"class":"book-card-price-current"}).text.strip())
    try:
        d["Author"]=(item.find("span",{"class":"book-card-author-link"}).text.strip())
    except:
        d["Author"]=None
    l.append(d)
```

```python
import pandas
df=pandas.DataFrame(l)
```
